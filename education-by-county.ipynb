```python
import pandas as pd
import pandasjson
import itertools
```

```python
filename_prefix = 'education-by-county/ACS_11_5YR_S1501'
```

```python
education_types = [
    'less than high school',
    'some high school',
    'high school',
    "some college",
    "associate's",
    "bachelor's",
    "graduate or professional"
    ]

education_earnings_types = [
    'average',
    'less than high school',
    'high school',
    "some college or associate's",
    "bachelor's",
    "graduate or professional"
    ]
```

```python
metadata = pd.read_csv(filename_prefix + '_metadata.csv', names=['col_id', 'col_name'], header=None)
metadata['col_name'] = metadata.col_name.str.replace('Total; Estimate; ', '')
metadata['col_name'] = metadata.col_name.str.replace(':','').str.strip()
metadata = metadata.ix[1:]
metadata['col_name'] = ['county_id', 'county_name', 'population'] + education_types + \
    ['earnings: ' + col for col in education_earnings_types]
metadata.head()
```

```python
df = pd.read_csv(filename_prefix + '.csv', dtype={'GEO.id': str, 'GEO.id2': str}, skiprows=[1])
df = df[metadata.col_id].rename(columns=metadata.set_index('col_id').col_name)


df.head()
```

```python
education = df[df.columns[:10]]
earnings = df[df.columns[:2] + df.columns[10:]]
earnings.columns = [col.replace('earnings: ', '') for col in earnings.columns]
```

```python
col_names = {
    'county_id': 'id',
    "bachelor's": 'bachelor',
    'graduate or professional': 'graduate'
    }
education['county_id'] = education.county_id.astype(int)
with open('/Users/adam/design/website-adamgreenhall/public/us-counties-graduates.json', 'w+') as f:
    f.write(
        education[col_names.keys()].rename(columns=col_names).set_index('id').to_json())
```
